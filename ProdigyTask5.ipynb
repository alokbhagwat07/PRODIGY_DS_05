```python
# Fillinng missing values...
dataset.fillna(dataset.mode().iloc[0], inplace=True)
dataset
```

```python
# No of columns in dataset
dataset.columns
```

```python
# Now not null values dataset is clean....
dataset.isnull().sum()
```

```python
# Drop unnecessarily columns....
dataset=dataset.drop(columns=['Service_year_of_vehicle'])
dataset
```

```python
# Converting time into hour....
dataset["Hour"]=pd.to_datetime(dataset["Time"], format = "%H:%M:%S").dt.hour
```

```python
# Create time day function ...
def time_category(hour):
    if 5<= hour < 12 :
        return "Morning"
    elif 12<= hour < 17:
        return "Afternoon" 
    elif 17<= hour < 21 :
        return "Evening"
    else:
        return "Night"
dataset["Time_Day"]=dataset["Hour"].apply(time_category)
```

```python
# It Works Successfully....
print(dataset["Time_Day"])
```

```python
# Now Analysis of accident on Time And Day...
sns.countplot(x="Time_Day",data=dataset)
plt.title("Accidents by Time Of Day...")
plt.show()
```

Most accidents occur during Evening and Afternoon, indicating high traffic density...

```python
# Accidents due to whether condition...
plt.figure(figsize=(10,5))
dataset["Weather_conditions"].value_counts().plot(kind="bar")
plt.title("Accidents Due To Whether Conditions....")
plt.show()
```

It Shows at normal conditions more number of accidents are happened...

```python
# Gennder Category
sns.countplot(x='Sex_of_driver',data=dataset)
plt.title("Male And Female No Of Accidents ")
plt.show()
```

It shows Approx Male Accidents is extremely large as compared to female...

```python
# Road Surface Condition vs Accident Severity
sns.countplot(x="Road_surface_conditions", hue="Accident_severity",data=dataset)
plt.title("Road Surface vs Accident Severity")
plt.xticks(rotation=45)
plt.show()
```

It shows Dry Roads have maximum number of Accident Severnity...

```python
# Light Condition vs Accident Severity
sns.countplot(x="Light_conditions", hue="Accident_severity",data=dataset)
plt.title("Light Conditions  vs Accident Severity")
plt.xticks(rotation=45)
plt.show()
```

It Shows Most Accidents Occur At Day Time and most is there is little amount of light...
